```python
import pandas as pd 
import seaborn as sns 
import folium
```

```python
airbnb_dat = pd.read_csv("/Users/jameskulp/Desktop/DS 2002/airbnb_dataset.csv") 
airbnb_dat.head()
```

```python
#Histogram of Price 
sns.histplot(airbnb_dat['Price'], bins = 10)
```

```python
#Histogram of Rating 
sns.histplot(airbnb_dat['Rating'], bins = 10)
```

```python
#Scatter plot of Price and Rating 
sns.scatterplot(data=airbnb_dat, x = 'Price', y = 'Rating')
```

```python
#Mapping with Folium 
map_airbnb = folium.Map(location=[40.7, -73.85], zoom_start=12)

# Add markers for each Airbnb listing
for _, row in airbnb_dat.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"Price: ${row['Price']} | Rating: {row['Rating']:.1f}"
    ).add_to(map_airbnb)

# Display the map
map_airbnb
```
